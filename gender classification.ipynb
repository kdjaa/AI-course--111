```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x = df[['long_hair', 'forehead_width_cm', 'forehead_height_cm', 'nose_wide',
       'nose_long', 'lips_thin', 'distance_nose_to_lip_long']]

y= df['gender']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y,test_size = 0.2, random_state = 42)
```

```python
len(x_train)
```

```python
len(y_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(f"accuracy score", accuracy)
print("-----------------")
print(f"classification report\n", class_report)
print("-----------------")
print(f"confusion matrix\n", conf_matrix)
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
```

```python
svm_model.fit(x_train, y_train)
```

```python
svm_y_pred = svm_model.predict(x_test)
```

```python
svm_accuracy = accuracy_score(y_test, svm_y_pred)
svm_class_report = classification_report(y_test, svm_y_pred)
svm_conf_matrix = confusion_matrix(y_test, svm_y_pred)
```

```python
print(f"accuracy score", svm_accuracy)
print("-----------------")
print(f"classification report\n", svm_class_report)
print("-----------------")
print(f"confusion matrix\n", svm_conf_matrix)
```

```python
print(f"accuracy score after scaling", svm_accuracy_2)
print("-----------------")
print(f"classification report after scaling\n", svm_class_report_2)
print("-----------------")
print(f"confusion matrix after scaling\n", svm_conf_matrix_2)
```
