```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.neighbors import KNeighborsClassifier
```

```python
df.columns
```

```python
df['income'].unique()
```

```python
df['income'] = df['income'].replace({' <=50K':0,' >50K':1 })
```

```python
df['income'].value_counts()
```

```python
x = df[['age', 'fnlwgt','education_num','sex','capital_gain', 'capital_loss', 'hours_per_week',]]

y= df['income']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
len(x_train)
```

```python
y_test.shape
```

```python
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
y_pred = logistic_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test,y_pred)
```

```python
print(f"accuracy score", accuracy)
print("-----------------")
print(f"classification report\n", class_report)
print("-----------------")
print(f"confusion matrix\n", conf_matrix)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model_after_scaling = LogisticRegression()
```

```python
model_after_scaling.fit(x_train_scaled,y_train)
```

```python
y_predic = model_after_scaling.predict(x_test_scaled)
```

```python
accuracy_2 = accuracy_score(y_test,y_predic)
```

```python
conf_matrix_2 = confusion_matrix(y_test, y_predic)
class_report_2 = classification_report(y_test,y_predic)
```

```python
print(f"accuracy score", accuracy_2)
print("-----------------")
print(f"classification report\n", class_report_2)
print("-----------------")
print(f"confusion matrix\n", conf_matrix_2)
```

```python
svm_model = SVC()
```

```python
svm_model.fit(x_train_scaled,y_train)
```

```python
svm_y_pred = svm_model.predict(x_test_scaled)
```

```python
svm_accuracy = accuracy_score(y_test, svm_y_pred)
svm_conf_matrix = confusion_matrix(y_test,svm_y_pred )
svm_class_report = classification_report(y_test,svm_y_pred)
```

```python
print(f"accuracy score",svm_accuracy)
print("-----------------")
print(f"classification report\n",svm_class_report)
print("-----------------")
print(f"confusion matrix\n", svm_conf_matrix)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
```
